```python
# Install additional packages required for API integration and web scraping
!pip install imdbpy justwatch-scraper beautifulsoup4 requests --quiet

# Import necessary libraries
import requests
from bs4 import BeautifulSoup
from imdb import IMDb
from justwatch import JustWatch
import pandas as pd

# Initialize IMDb and JustWatch clients
imdb_client = IMDb()
justwatch_client = JustWatch(country='US')

# Function to fetch data from IMDb API
def fetch_imdb_data(title):
    """Fetch movie or TV show data from IMDb."""
    try:
        search_results = imdb_client.search_movie(title)
        if search_results:
            movie = search_results[0]
            imdb_client.update(movie)
            return {
                "Title": movie.get('title'),
                "Release Year": movie.get('year'),
                "IMDb Score": movie.get('rating'),
                "Number of IMDb Votes": movie.get('votes'),
                "Director/Creator": ', '.join([director['name'] for director in movie.get('directors', [])]),
                "Cast": ', '.join([actor['name'] for actor in movie.get('cast', [])[:5]]),  # Limit to top 5 cast members
                "Description/Synopsis": movie.get('plot outline'),
                "Age Rating": movie.get('certificates')[0] if movie.get('certificates') else None
            }
    except Exception as e:
        print(f"Error fetching IMDb data for {title}: {e}")
    return None

# Function to fetch data from JustWatch API
def fetch_justwatch_data(title):
    """Fetch availability and additional data from JustWatch."""
    try:
        results = justwatch_client.search_for_item(query=title)
        if results['items']:
            item = results['items'][0]
            return {
                "Title": item.get('title'),
                "Type": item.get('object_type'),
                "Amazon Prime URL": next((offer['urls']['standard_web'] for offer in item['offers'] if offer['provider_id'] == 9), None),
                "Is Available": any(offer['provider_id'] == 9 for offer in item['offers']),
                "Genre(s)": ', '.join(item.get('genre_ids', [])),
                "Language": item.get('original_language'),
                "Country of origin": item.get('production_countries', [])[0] if item.get('production_countries') else None
            }
    except Exception as e:
        print(f"Error fetching JustWatch data for {title}: {e}")
    return None

# Sample data collection
sample_titles = ["The Boys", "The Marvelous Mrs. Maisel", "Jack Ryan"]

# Collect data for sample titles
collected_data = []
for title in sample_titles:
    imdb_data = fetch_imdb_data(title)
    justwatch_data = fetch_justwatch_data(title)
    if imdb_data and justwatch_data:
        combined_data = {**imdb_data, **justwatch_data}
        collected_data.append(combined_data)

# Convert collected data to DataFrame
sample_df = pd.DataFrame(collected_data)

# Display the sample data
print(sample_df.head())

# Save the sample data to a CSV file
sample_csv_path = '/home/user/amazon_prime_dataset/sample_amazon_prime_data.csv'
sample_df.to_csv(sample_csv_path, index=False, encoding='utf-8')

print(f"Sample data collection complete. Data saved to {sample_csv_path}.")
```
